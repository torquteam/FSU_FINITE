```python
# Figure 4 (Single TOV solve plot)
import matplotlib.pyplot as plt
import numpy as np

# Set font to a more professional style
plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
    'font.serif': ['Computer Modern'],
    'font.size': 14,
    'axes.titlesize': 14,
    'axes.labelsize': 16,
    'xtick.labelsize': 15,
    'ytick.labelsize': 15,
    'legend.fontsize': 14,
    'text.latex.preamble': r'\usepackage{bm}',
})

# Data for plotting left
garnet = np.loadtxt("FSUGarnet_Core_EOS.txt")
gold = np.loadtxt("FSUGold2_Core_EOS.txt")

# Data for plotting right
gold_comp = np.loadtxt("FSUGold2_comp.txt")
garnet_comp = np.loadtxt("FSUGarnet_comp.txt")


# Create figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(gold[:,0], gold[:,2], color='black',lw=1.5,label='FSUGold2')
ax[0].plot(garnet[:,0], garnet[:,2], color='maroon',lw=1.5,label='FSUGarnet')

ax[1].plot(gold[:,0], gold_comp[:,0], color='black',lw=1.5)
ax[1].plot(gold[:,0], gold_comp[:,1], color='black',lw=1.5, ls='dashed')
ax[1].plot(gold[:,0], gold_comp[:,2], color='black',lw=1.5, ls='dotted')
ax[1].plot(gold[:,0], gold_comp[:,3], color='black',lw=1.5, ls='-.')

ax[1].plot(garnet[:,0], garnet_comp[:,0], color='maroon',lw=1.5,label='n')
ax[1].plot(garnet[:,0], garnet_comp[:,1], color='maroon',lw=1.5,ls='dashed',label='p')
ax[1].plot(garnet[:,0], garnet_comp[:,2], color='maroon',lw=1.5,ls='dotted',label=r'$e^-$')
ax[1].plot(garnet[:,0], garnet_comp[:,3], color='maroon',lw=1.5,ls='-.',label=r'$\mu^-$')

    
# Customization for left plot
ax[0].set_xlabel(r'$\rho \,\,(\textrm{fm}^{-3})$',fontweight='bold')
ax[0].set_ylabel(r'$P \,\, (\textrm{MeV}/\textrm{fm}^{-3})$',fontweight='bold')
ax[0].set_yscale('log')
ax[0].legend(loc='lower right')

# Customization for right plot
ax[1].set_xlabel(r'$\rho \,\, (\textrm{fm}^{-3})$',fontweight='bold')
ax[1].set_ylabel('Relative Population',fontweight='bold')
ax[1].set_yscale('log')
ax[1].legend(loc='lower right')

# Adjust layout
plt.tight_layout()

# Save the figure (optional)
plt.savefig('Figure4.png', dpi=500)
```

```python
# Figure 5 (Single TOV solve plot)
import matplotlib.pyplot as plt
import numpy as np

# Set font to a more professional style
plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
    'font.serif': ['Computer Modern'],
    'font.size': 14,
    'axes.titlesize': 14,
    'axes.labelsize': 16,
    'xtick.labelsize': 15,
    'ytick.labelsize': 15,
    'legend.fontsize': 14,
    'text.latex.preamble': r'\usepackage{bm}',
})

# Data for plotting left
garnetCa = np.loadtxt("densities48,20_gar.txt")
goldCa = np.loadtxt("densities48,20_gold.txt")
expCa = np.loadtxt("FBDens48Ca.txt")

# Data for plotting right
garnetPb = np.loadtxt("densities208,82_gar.txt")
goldPb = np.loadtxt("densities208,82_gold.txt")
expPb = np.loadtxt("FBDens208Pb.txt")


# Create figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].plot(garnetCa[:,0], garnetCa[:,7], color='maroon',lw=1.5,label='FSUGarnet')
ax[0].plot(goldCa[:,0], goldCa[:,7], color='navy',lw=1.5,label='FSUGold2')
ax[0].plot(expCa[:,0], expCa[:,1], color='black',lw=1.5,label='Exp',ls='dashed')

ax[1].plot(garnetPb[:,0], garnetPb[:,7], color='maroon',lw=1.5,label='FSUGarnet')
ax[1].plot(goldPb[:,0], goldPb[:,7], color='navy',lw=1.5,label='FSUGold2')
ax[1].plot(expPb[:,0], expPb[:,1], color='black',lw=1.5,label='Exp',ls='dashed')

    
# Customization for left plot
ax[0].set_xlabel('r (fm)',fontweight='bold')
ax[0].set_ylabel(r'$\rho_\textrm{ch} \, \, (\textrm{fm}^{-3})$',fontweight='bold')
ax[0].legend(loc='lower left')
ax[0].set_xlim(0,7)
ax[0].text(5.5,0.08,r'$^{48}$\textrm{Ca}',fontsize=16)

# Customization for right plot
ax[1].set_xlabel('r (fm)',fontweight='bold')
ax[1].set_ylabel(r'$\rho_\textrm{ch} \, \, (\textrm{fm}^{-3})$',fontweight='bold')
ax[1].set_xlim(0,10)
ax[1].text(7.5,0.0625,r'$^{208}$\textrm{Pb}',fontsize=16)

# Adjust layout
plt.tight_layout()

# Save the figure (optional)
plt.savefig('Figure5.png', dpi=500)
```
