```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({
    'text.usetex': True,
    'font.family': 'serif',
    'font.serif': ['Computer Modern'],
    'font.size': 20,
    'axes.titlesize': 14,
    'axes.labelsize': 21,
    'xtick.labelsize': 21,
    'ytick.labelsize': 21,
    'legend.fontsize': 16,
    'text.latex.preamble': r'\usepackage{bm}',
})

DINOa_Symm = np.loadtxt("DINOa_Symm.txt")
FSUTDb_Symm = np.loadtxt("FSU-TDb_Symm.txt")
FSUT_Symm = np.loadtxt("FSU-T_Symm.txt")

# Adjust start and end of EOS data
l = 0
m = 170

# plot XEFT data and FSU PNM EOS
plt.figure(figsize=(8,8))

plt.plot(DINOa_Symm[l:m,0],DINOa_Symm[l:m,2],color='blue',linewidth=3)
plt.plot(FSUT_Symm[l:m,0],FSUT_Symm[l:m,2],color='green',linewidth=3)
plt.plot(FSUTDb_Symm[l:m,0],FSUTDb_Symm[l:m,2],color='red',linewidth=3)
plt.legend(['DINOa','FSU-TD0','FSU-TD2'],fontsize=21)

plt.plot(DINOa_Symm[l:m,0],DINOa_Symm[l:m,1],color='blue',linewidth=3,linestyle='dashed')
plt.plot(FSUT_Symm[l:m,0],FSUT_Symm[l:m,1],color='green',linewidth=3,linestyle='dashed')
plt.plot(FSUTDb_Symm[l:m,0],FSUTDb_Symm[l:m,1],color='red',linewidth=3,linestyle='dashed')

plt.xlabel(r'$\rho$ ($\rm fm^{-3}$)')
plt.ylabel('Symmetry Energy (MeV)')
plt.rc('axes', labelsize=30)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=22)    # fontsize of the tick labels
plt.rc('ytick', labelsize=22)    # fontsize of the tick labels
plt.tight_layout()

plt.savefig('Figure25',dpi=500)
```
