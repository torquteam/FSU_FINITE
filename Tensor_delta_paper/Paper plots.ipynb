```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# import MR data
dinoa_RM = np.loadtxt("DINOa_RM.txt")
plambda = np.loadtxt("DINOa+scalar_RM.txt")
nlambda = np.loadtxt("DINOa-scalar_RM.txt")


# Generate 67 and 95 percent contours
##############################################
def eigsorted(cov):
    vals, vecs = np.linalg.eigh(cov)
    order = vals.argsort()[::-1]
    return vals[order], vecs[:,order]

# Define the mean and covariance matrix for the bivariate Gaussian
mean = [0.041, 0.0277]  # Mean values for x and y
covariance_matrix = [[0.013**2, 0], [0, 0.0055**2]]  # Covariance matrix

# Number of data points
num_points = 10000

# Generate x and y values from the bivariate Gaussian distribution
x, y = np.random.multivariate_normal(mean, covariance_matrix, num_points).T

# plot XEFT data and FSU PNM EOS
fig, ax1 = plt.subplots(1,1,figsize=(8,8))
# Plot MR curves
ax1.plot(dinoa_RM[:,3],dinoa_RM[:,4],color='black',lw=3)
ax1.plot(plambda[:,3],plambda[:,4],color='green',lw=3)
ax1.plot(nlambda[:,3],nlambda[:,4],color='red',lw=3)

ax1.set(xlabel='Radius (km)',ylabel=r'Mass ($\rm M/M_{\odot}$)')

ax1.legend([r'DINOa $(\Lambda_{1.4} = 1326.1)$',r"DINOa + $\Lambda_s$  $(\Lambda_{1.4} = 1191.7$)",r"DINOa - $\Lambda_s$  $(\Lambda_{1.4} = 1511.9$)"],loc=[0.41,0.19],fontsize=18)
ax1.set_ylim([0,2.3])
ax1.set_xlim([12,19])

left, bottom, width, height = [0.58, 0.57, 0.36, 0.36]
ax2 = fig.add_axes([left, bottom, width, height])

# plot the contours
#sqrt[-2*ln(1-p)]
nstd = 1.489
cov = np.cov(x, y)
vals, vecs = eigsorted(cov)
theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
w, h = 2 * nstd * np.sqrt(vals)
ell = Ellipse(xy=(np.mean(x), np.mean(y)),
              width=w, height=h,
              angle=theta, color='black',lw=2)
ell.set_facecolor('none')
ax2.add_artist(ell)

nstd = 2.4475
cov = np.cov(x, y)
vals, vecs = eigsorted(cov)
theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))
w, h = 2 * nstd * np.sqrt(vals)
ell = Ellipse(xy=(np.mean(x), np.mean(y)),
              width=w, height=h,
              angle=theta, color='black',lw=2)
ell.set_facecolor('none')
ax2.add_artist(ell)

# Plot DINO models weak skin
ax2.plot(0.0263,0.0333,marker='s',markerfacecolor='black',markeredgecolor='black',markersize=10,lw=3) #DINO a
ax2.plot(0.0303,0.0349,marker='s',markerfacecolor='black',markeredgecolor='black',markersize=10,lw=3) #DINO b
ax2.plot(0.0344,0.0367,marker='s',markerfacecolor='black',markeredgecolor='black',markersize=10,lw=3) #DINO c

# Plot DINO + lambda_s weak skin
ax2.plot(0.02929,0.0300,marker='d',markerfacecolor='red',markeredgecolor='black',markersize=14,lw=3) #DINO a - lambda_s (-0.00025)
ax2.plot(0.02585,0.03469,marker='d',markerfacecolor='green',markeredgecolor='black',markersize=14,lw=3) #DINO a + lambda_s (0.00025)

# Set the x-axis and y-axis limits for weak skin
ax2.set_xlim(0.02, 0.04)  # Set x-axis limits from 0 to 6
ax2.set_ylim(0.02, 0.05)  # Set y-axis limits from 0 to 6
ax2.set(xlabel=r'${\rm F}_{\rm ch} - {\rm F}_{\rm wk}$ ($^{208}$Pb)',ylabel=r'${\rm F}_{\rm ch} - {\rm F}_{\rm wk}$ ($^{48}$Ca)')
ax2.annotate(r'67%', xy=(0.022, 0.033), xytext=(0.022, 0.033),color='black',size=16)
ax2.annotate(r'95%', xy=(0.022, 0.040), xytext=(0.022, 0.040),color='black',size=16)
#ax2.text(0.024,0.024,'67%',fontsize=15)
#ax2.text(0.014,0.021,'95%',fontsize=15)

plt.rc('axes', labelsize=25)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=25)    # fontsize of the tick labels
plt.rc('ytick', labelsize=25)    # fontsize of the tick labels
ax2.tick_params(axis='both', which='both', labelsize=16)
ax2.xaxis.label.set_size(20)  # Adjust x-axis label size
ax2.yaxis.label.set_size(20)  # Adjust y-axis label size

plt.tight_layout()
plt.savefig('Figure3.png',transparent=False,facecolor='white')
```

```python
import matplotlib.pyplot as plt
import numpy as np

densityPb_dinoa_n = np.loadtxt("Pb208-DINOa-N.txt")
densityPb_dinoa_p = np.loadtxt("Pb208-DINOa-P.txt")
densityPb_dinob_n = np.loadtxt("Pb208-DINOb-N.txt")
densityPb_dinob_p = np.loadtxt("Pb208-DINOb-P.txt")
densityPb_dinoc_n = np.loadtxt("Pb208-DINOc-N.txt")
densityPb_dinoc_p = np.loadtxt("Pb208-DINOc-P.txt")
densityPb_td0_n = np.loadtxt("Pb208-FSU-TD0-N.txt")
densityPb_td0_p = np.loadtxt("Pb208-FSU-TD0-P.txt")

fig, axs = plt.subplots(1, 1,figsize=(8,6),sharey=False, sharex=False, facecolor='white')

#neutron point densities
################################################
r = densityPb_dinoa_n[:,0]
the = densityPb_dinoa_n[:,1]
axs.plot(r, the, color='red',ls='solid',lw=3)

r = densityPb_dinob_n[:,0]
the = densityPb_dinob_n[:,1]
axs.plot(r, the, color='blue',ls='solid',lw=3)

r = densityPb_dinoc_n[:,0]
the = densityPb_dinoc_n[:,1]
axs.plot(r, the, color='green',ls='solid',lw=3)

r = densityPb_td0_n[:,0]
the = densityPb_td0_n[:,1]
axs.plot(r, the, color='black',ls='solid',lw=3)
###################################################

r = densityPb_dinoa_p[:,0]
the = densityPb_dinoa_p[:,1]
axs.plot(r, the, color='red',ls='dashed',lw=3)

r = densityPb_dinob_p[:,0]
the = densityPb_dinob_p[:,1]
axs.plot(r, the, color='blue',ls='dashed',lw=3)

r = densityPb_dinoc_p[:,0]
the = densityPb_dinoc_p[:,1]
axs.plot(r, the, color='green',ls='dashed',lw=3)

r = densityPb_td0_p[:,0]
the = densityPb_td0_p[:,1]
axs.plot(r, the, color='black',ls='dashed',lw=3)

plt.xlabel("r (fm)",size=20)
axs.set_ylabel(r"$\rm \rho(r)$ (${\rm fm}^{-3}$) ",size=20)

leg = axs.legend(["DINOa","DINOb","DINOc","FSU-TD0"],fontsize=20,loc=[0.65,0.4])

axs.text(3.15,0.045,'Protons',fontsize=19)
axs.text(3,0.10,'Neutrons',fontsize=19)
axs.text(8,0.14,r'$^{208}$Pb',fontsize=28)

axs.set_xlim(0,10)
#axs.set_ylim(0,0.12)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
plt.savefig('Figure0')
```
