```python
import corner
import numpy as np
import matplotlib.pyplot as plt
import statistics
samples = np.loadtxt("MCMC_complete.txt")

mw = 782.5
mp = 763.0
couplings = []
indicies = [8,9,10,12,13,14,16]
labels = ["ms","mw","mp","gs2","gw2","gp2","kappa","lambda","zeta","lambda_v"]

mean = statistics.mean(samples[:,7])
couplings.append(mean)
couplings.append(mw)
couplings.append(mp)
for i in indicies:
    mean = statistics.mean(samples[:,i])
    couplings.append(mean)

couplings[3] = couplings[3]*couplings[0]**2
couplings[4] = couplings[4]*mw**2
couplings[5] = couplings[5]*mp**2
print(couplings)

with open("MTOV28_raw.txt",'w') as output:
    print("#central values for couplings",file=output)
    for i in range(10):
        print(f"{labels[i]} = {couplings[i]}",file=output)
    print("",file=output)
    for i in range(len(samples)):
        print(samples[i,18],file=output)
```
