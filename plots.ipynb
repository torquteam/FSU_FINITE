```python
import numpy as np
from scipy.integrate import simpson

# Define a function to read data from a text file
def read_data_from_file(file_path):
    data = np.loadtxt(file_path)
    return data

# Define a function to calculate the integral of the data
def calculate_integral(data):
    x = data[:, 0]  # Assuming the first column contains x values
    y = data[:, 1]  # Assuming the second column contains y values
    integral = simpson(y, x)
    return integral

# Replace 'your_data.txt' with the actual file path of your data
file_path = "en_integral.txt"

try:
    data = read_data_from_file(file_path)
    integral = calculate_integral(data)
    energy = ((2*np.pi*integral+3302.656403)*13.2695841 - 0.75*41*48**(-1/3))/48 - 939
    print(f"Integral of the data: {integral}")
    print(f"Energy is: {energy}")
except FileNotFoundError:
    print(f"File not found: {file_path}")
except Exception as e:
    print(f"An error occurred: {str(e)}")
```

```python
import numpy as np
from scipy.interpolate import CubicSpline
integ = np.loadtxt("integrand.txt")

def simpsons(a, fa, b, fb):
    """Evaluates the Simpson's Rule, also returning m and f(m) to reuse"""
    m = (a + b) / 2
    fm = cs(m)
    return (m, fm, abs(b - a) / 6 * (fa + 4 * fm + fb))

def adaptive_simpsons(a, fa, b, fb, eps, whole, m, fm):
    """
    Efficient recursive implementation of adaptive Simpson's rule.
    Function values at the start, middle, end of the intervals are retained.
    """
    lm, flm, left  = simpsons(a, fa, m, fm)
    rm, frm, right = simpsons(m, fm, b, fb)
    delta = 0.1*abs(whole-left-right)
    if delta < eps:
        return left + right 
    return adaptive_simpsons(a, fa, m, fm, eps/2, left , lm, flm) +\
           adaptive_simpsons(m, fm, b, fb, eps/2, right, rm, frm)

def integrate_array(array, eps):
    global x, y, cs
    x, y = array[:, 0], array[:, 1]
    cs = CubicSpline(x, y)  # Create cubic spline interpolator
    """Integrate f from a to b using Adaptive Simpson's Rule with max error of eps."""
    a, b = x[0], x[-1]
    fa, fb = y[0], y[-1]
    m, fm, whole = simpsons(a, fa, b, fb)
    return adaptive_simpsons(a, fa, b, fb, eps, whole, m, fm)

# Example usage:
# Replace this array with your own data
array = integ

result = integrate_array(array,1e-8)
print(f"The integrated value is: {result}")

integral = simpson(y, x,axis=0)
print(integral)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import diff

en = np.loadtxt("S_alpha.txt")
alpha = en[:,0]
en = en[:,1]

# Fit a polynomial to the data
degree = 18  # Degree of the polynomial to fit
coefficients = np.polyfit(alpha, en, degree)
p = np.poly1d(coefficients)

dc2 = np.polyder(coefficients,2)
dp2 = np.poly1d(dc2)

dc4 = np.polyder(coefficients,4)
dp4 = np.poly1d(dc4)

en1 = np.gradient(en, alpha,edge_order=2)
en2 = np.gradient(en1, alpha,edge_order=2)
en3 = np.gradient(en2, alpha,edge_order=2)
en4 = np.gradient(en3, alpha, edge_order=2)

fig, axs = plt.subplots(1, 1,figsize=(10,6),sharey=False, sharex=False, facecolor='white')
axs.plot(alpha,en, color='green',ls='solid')
axs.plot(alpha,p(alpha), color='red',ls='solid')
axs.plot(alpha,dp2(alpha), color='red',ls='dashed')
axs.plot(alpha,dp4(alpha), color='red',ls='dotted')
print(dp2(0))
print(dp4(0))
#axs.plot(alpha,en2, color='green',ls='dashed')
axs.plot(alpha,en4, color='red',ls='solid')
axs.set_xlim(0.02,0.8)
axs.set_ylim(-10000,20000)

################################################
plt.xlabel("alpha",size=20)
axs.set_ylabel("E/A",size=20)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define your array
x = alpha
y = en  # Example array, replace it with your own data

# Calculate the fourth derivative using numpy gradient
fourth_derivative = np.gradient(np.gradient(np.gradient(np.gradient(y, x), x), x), x)

# Apply moving average for smoothing
window_size = 5
smoothed_fourth_derivative = np.convolve(fourth_derivative, np.ones(window_size)/window_size, mode='valid')

# Adjust the x values for the smoothed derivative
smoothed_x = x[:-window_size+1]

# Plot the original fourth derivative and the smoothed version
plt.figure(figsize=(10, 8))
#plt.plot(x, fourth_derivative, label='Original 4th Derivative', alpha=0.5)
plt.plot(smoothed_x, smoothed_fourth_derivative, label='Smoothed 4th Derivative (Moving Average)', linestyle='--')
plt.legend()
plt.title('Smoothing the Fourth Derivative')
plt.xlabel('x')
plt.ylabel('Fourth Derivative')
plt.xlim(0.02, 0.8)  # Set x-axis limits
plt.ylim(0, 10000)  # Set x-axis limits
plt.grid(True)
plt.show()
```
