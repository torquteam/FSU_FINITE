```python
import numpy as np
import matplotlib.pyplot as plt
ms = np.loadtxt("ms.txt")

# Example x and y data points
x = ms[:,0]
y = np.log(ms[:,1])

# Fit a polynomial of degree n (choose an appropriate degree)
degree = 2
coefficients = np.polyfit(x, y, degree)
polynomial = np.poly1d(coefficients)

# Derivative of the polynomial
derivative = polynomial.deriv()

# Find the roots of the derivative (critical points)
critical_points = derivative.r
print(critical_points)
# Evaluate the polynomial at the critical points
critical_values = polynomial(critical_points)

# Find the maximum value among the critical points
max_value = np.max(critical_values)
print(critical_values)
max_point = critical_points[np.argmax(critical_values)]

print(f"Maximum value: {np.exp(max_value)} at x = {max_point}")

# Plotting the polynomial and the maximum point
x_vals = np.linspace(min(x), max(x), 1000)
y_vals = polynomial(x_vals)

plt.plot(x, y, 'o', label='Data points')
plt.plot(x_vals, y_vals, '-', label=f'Fitted polynomial (degree {degree})')
plt.plot(max_point, max_value, 'ro', label='Maximum point')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
